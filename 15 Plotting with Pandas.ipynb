```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminder2007.csv')
df.head()
```

```python
plt.figure(figsize=(12, 6))
for continent in df['continent'].unique():
    data = df[df['continent'] == continent]
    plt.plot(data['gdpPercap'], data['lifeExp'], 'o', label=continent)
plt.xlabel('GDP per capita')
plt.ylabel('Life expectancy')
plt.title('Life expectancy vs. GDP per capita')
plt.legend()
plt.show()
```

```python


plt.figure(figsize=(12, 6))
for continent in df['continent'].unique():
    data = df[df['continent'] == continent] 
    plt.plot(np.arange(len(data)), data['gdpPercap'], '', label=continent)
plt.xlabel('Year')
plt.ylabel('GDP per capita')
plt.title('GDP per capita over time')
plt.legend()
plt.show()
```

```python
# histogram
plt.figure(figsize=(12, 6))
for continent in df['continent'].unique():
    data = df[df['continent'] == continent]
    plt.hist(data['lifeExp'], alpha=0.5, label=continent)
plt.xlabel('Life expectancy')
plt.ylabel('Frequency')
plt.title('Distribution of life expectancy')
plt.legend()
plt.show()
```
